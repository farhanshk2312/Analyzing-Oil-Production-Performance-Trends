```python
import pandas as pd

# Load your wide-format CSV
df = pd.read_csv("PET_CRD_CRPDN_ADC_MBBL_M.csv")

# Convert first column to datetime and rename it to 'Date'
# df.iloc[:, 0] = pd.to_datetime(df.iloc[:, 0], format="%m/%d/%y")
# df.rename(columns={df.columns[0]: "Date"}, inplace=True)

# Melt the dataframe from wide to long format
df_long = df.melt(
    id_vars=["Date"],
    var_name="Area",
    value_name="Volume"
)

# Define your hierarchy map
hierarchy_map = {
    # ----- PADD 1 -----
    "Florida": ("East Coast (PADD 1)", "State"),
    "New York": ("East Coast (PADD 1)", "State"),
    "Pennsylvania": ("East Coast (PADD 1)", "State"),
    "Virginia": ("East Coast (PADD 1)", "State"),
    "West Virginia": ("East Coast (PADD 1)", "State"),

    # ----- PADD 2 -----
    "Illinois": ("Midwest (PADD 2)", "State"),
    "Indiana": ("Midwest (PADD 2)", "State"),
    "Kansas": ("Midwest (PADD 2)", "State"),
    "Kentucky": ("Midwest (PADD 2)", "State"),
    "Michigan": ("Midwest (PADD 2)", "State"),
    "Missouri": ("Midwest (PADD 2)", "State"),
    "Nebraska": ("Midwest (PADD 2)", "State"),
    "North Dakota": ("Midwest (PADD 2)", "State"),
    "Ohio": ("Midwest (PADD 2)", "State"),
    "Oklahoma": ("Midwest (PADD 2)", "State"),
    "South Dakota": ("Midwest (PADD 2)", "State"),
    "Tennessee": ("Midwest (PADD 2)", "State"),

    # ----- PADD 3 -----
    "Alabama": ("Gulf Coast (PADD 3)", "State"),
    "Arkansas": ("Gulf Coast (PADD 3)", "State"),
    "Louisiana": ("Gulf Coast (PADD 3)", "State"),
    "Mississippi": ("Gulf Coast (PADD 3)", "State"),
    "New Mexico": ("Gulf Coast (PADD 3)", "State"),
    "Texas": ("Gulf Coast (PADD 3)", "State"),
    "Federal Offshore--Gulf of America": ("Gulf Coast (PADD 3)", "State"),

    # ----- PADD 4 -----
    "Colorado": ("Rocky Mountain (PADD 4)", "State"),
    "Idaho": ("Rocky Mountain (PADD 4)", "State"),
    "Montana": ("Rocky Mountain (PADD 4)", "State"),
    "Utah": ("Rocky Mountain (PADD 4)", "State"),
    "Wyoming": ("Rocky Mountain (PADD 4)", "State"),

    # ----- PADD 5 -----
    "Alaska": ("West Coast (PADD 5)", "State"),
    "Arizona Field": ("West Coast (PADD 5)", "State"),
    "California": ("West Coast (PADD 5)", "State"),
    "Nevada": ("West Coast (PADD 5)", "State"),
    "Federal Offshore": ("West Coast (PADD 5)", "State"),
}

# Map hierarchy to new columns
df_long["PADD_Region"] = df_long["Area"].map(lambda x: hierarchy_map.get(x, ("Unknown","Unknown"))[0])
df_long["Category"]    = df_long["Area"].map(lambda x: hierarchy_map.get(x, ("Unknown","Unknown"))[1])

# Reorder columns for clarity
df_long = df_long[["Date", "PADD_Region", "Area", "Volume"]]

# Save to a new CSV file
df_long.to_csv("production_hierarchy.csv", index=False)

print("✅ Saved as production_hierarchy.csv")
```
